```python
import pandas as pd
import numpy as np

player_df = pd.read_csv('Game_Logs_Runningback.csv')
player_df.replace('--', np.nan, inplace=True)
player_df = player_df.drop(columns = ["Position", "Home or Away", "Game Date", "Games Played", "Games Started", "Longest Reception", "Longest Rushing Run", "Fumbles"])
```

```python
player_df = player_df[player_df['Season'] == "Regular Season"]
player_df = player_df.drop(columns = ["Season"])
player_df = player_df.sort_values(by=['Player Id', 'Year', 'Week'])
```

```python
valid = player_df[['Receptions', 'Rushing Yards']].notna().any(axis=1)
player_df = player_df[valid]
```

```python
valid = player_df[['Receptions', 'Rushing Yards']].notna().any(axis=1)  # Identify valid games
player_game_counts = player_df[valid].groupby(['Player Id', 'Year']).size()

# Filter players with less than 6 valid games in a season
valid_players = player_game_counts[player_game_counts > 5].index
player_df = player_df[player_df.set_index(['Player Id', 'Year']).index.isin(valid_players)]
```

```python
player_df.replace(np.nan, 0, inplace=True)
```
